# ICASSP 2021 Demo Page
This demo page is for the paper __Revisiting Onsets and Frames Model with Additive Attention__. High resolution figures and the audio samples for the transcription results can be found here.

Source code for the paper is available at https://github.com/KinWaiCheuk/ICASSP2021.github.io

## Model Archetecture
Left: Onsets and Frames model with an additive attention mechanism<br/>
Right: Linear model with an additive attention mechanism

For Onsets and Frames model, the attention mechanism attends to only one of the three features: $\boldsymbol{x_{\text{spec}}}$ or $\hat{\boldsymbol{y}}_{\text{onset}}$ or $\boldsymbol{\hat{y}_{\text{feat}}}$
![](demo/model.png)
